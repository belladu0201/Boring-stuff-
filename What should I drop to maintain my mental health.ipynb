# Import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format ='retina'
```

# Considering dropping one of the major or the minor, set up the dataset

```python
drop = pd.read_csv('which to drop.csv')
drop
```

```python
drop = drop.assign(
    code = drop['Course'].apply(lambda x: x.split('-')[0]))
drop
```

## getting the dept & code out

```python
drop = drop.assign(
    dept = drop['code'].apply(lambda dept: dept.split(' ')[0])
)
drop
```

# Dropping  DSC Minor?

```python
dsc = drop[drop['dept'] == 'DSC']
dsc
```

```python
dsc = dsc.groupby('code').agg(['mean'])
dsc = dsc['Study Hrs/wk']
```

```python
dsc25 = dsc.sort_values(by ='mean', ascending = False)
dsc25
```

```python
dsc25 = dsc25.reset_index()
dsc25 = dsc25.assign(
    number = dsc25['code'].apply(lambda x: x.split(' ')[1])
)
dsc25
```

```python
sns.set_style("darkgrid")
ax = sns.catplot(data = dsc25, x = dsc25['number'], y = 'mean', height = 5, aspect = 1.5)
plt.title('Hours spending on DSC courses')
plt.xlabel('Courses from DSC department')
plt.ylabel('Average Hours spending in each of the classes')
plt.show()
```

# DSC MAJOR AVERAGE STUDYING TIME

```python
sns.set_theme(style="whitegrid")
ax = sns.boxplot(x=dsc25["mean"])
```

## Conclusion: DSC30 takes a lot of time, maybe need to P/NP

# Dropping COGS? NOPE!

```python
cogs = drop[drop['dept'] == 'COGS']
cogs
```

```python
cogs = cogs.groupby('code').agg(['mean'])
cogs = cogs['Study Hrs/wk']
```

```python
cogs = cogs.reset_index()
cogs
```

```python
cog_class = pd.DataFrame().assign(
    course = cogs['code'],
    hour = cogs['mean']
)
```

```python
cog_class
```

```python
cog_class.hour
```

```python
cog_class = cog_class.assign(
    code = cog_class['course'].apply(lambda x: x.split(' ')[1])
)
cog_class
```

```python
sns.set_style("darkgrid")
ax = sns.catplot(data = cog_class, x = cog_class['code'], y = 'hour', height = 13, aspect = 3.5)
plt.title('Hours spending on COGS courses')
plt.xlabel('Courses from COGS department')
plt.ylabel('Average Hours spending in each of the classes')
plt.show()
```
